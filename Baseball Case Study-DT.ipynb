# DATA CLEANING

```python
# AS column 2B has minimum correlated value hence we are dropping it so it will not affect our model outcome.
df=df.drop(['2B'],axis=1)
```

```python
df
```

# REMOVING OUTLIERS

```python
import numpy as np
from scipy.stats import zscore
z=np.abs(zscore(df))
z.shape
```

```python
(z)
```

```python
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
print(df.shape)
print(df_new.shape)
```

# Percentage Data Loss

```python
loss_percent=(30-29)/30*100
print(loss_percent)
```

Dividing data into features and vectors

```python
x=df_new.iloc[:,:-1]
y=df_new.iloc[:,-1]
```

# Transforming the data to remove skewness

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```
