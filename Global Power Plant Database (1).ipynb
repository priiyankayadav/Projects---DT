# Data Pre-processing

```python
reg_df = df.copy()
```

```python
reg_df = pd.get_dummies(data=reg_df,columns=['primary_fuel'],prefix=['primary_fuel'],drop_first=True)
reg_df = pd.get_dummies(data=reg_df,columns=['geolocation_source'],prefix=['geolocation_source'],drop_first=True)
```

```python
reg_df.shape
```

# Train - Test Split

```python
from sklearn.model_selection import train_test_split
X = reg_df.drop('capacity_mw',axis=1)
y = reg_df['capacity_mw']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Linear Regression Model Implementation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()#Instantiate the object
```

```python
lr.fit(X_train,y_train)#call the fit method of linear regression class to train the model
```

```python
y_pred = lr.predict(X_test)
```

# Accuracy score for linear regression

```python
print("r2score::",r2_score(y_test,y_pred))
```

# Classification Model Pre-Processing

```python
classification_df = df.copy()
```

```python
classification_df = pd.get_dummies(data=classification_df,columns=['geolocation_source'],prefix=['geolocation_source'],drop_first=True)
```

```python
classification_df.columns
```

```python
X1 = classification_df.drop('primary_fuel',axis=1)
y1 = classification_df['primary_fuel']
```

```python
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size=0.20,random_state=1)
```

```python
print(X1_train.shape)
print(y1_train.shape)
print(X1_test.shape)
print(y1_test.shape)
```

# Classification Model Implementation

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')#Instantiate the object
dtc.fit(X1_train,y1_train)#call the fit method
```

```python
dtc_y_predict = dtc.predict(X1_test)
```

```python
from sklearn.metrics import accuracy_score
```

# Accuracy Score for Classification

```python
print("Accuracy score:",accuracy_score(y1_test,dtc_y_predict))#printing overall accuracy score
```

```python
from sklearn import metrics

# roc curve for classes
fpr = {}
tpr = {}
thresh ={}

n_class = 8
probs = dtc.predict_proba(X1_test)
#preds = probs[:,1]

for i in range(n_class):    
    fpr[i], tpr[i], thresh[i] = metrics.roc_curve(y1_test, probs[:,i], pos_label=i)
    
# plotting    
plt.plot(fpr[0], tpr[0], linestyle='--',color='orange', label='Class 0 vs Rest')
plt.plot(fpr[1], tpr[1], linestyle='--',color='green', label='Class 1 vs Rest')
plt.plot(fpr[2], tpr[2], linestyle='--',color='blue', label='Class 2 vs Rest')
plt.plot(fpr[3], tpr[3], linestyle='--',color='yellow', label='Class 3 vs Rest')
plt.plot(fpr[4], tpr[4], linestyle='--',color='cyan', label='Class 4 vs Rest')
plt.plot(fpr[5], tpr[5], linestyle='--',color='magenta', label='Class 5 vs Rest')
plt.plot(fpr[6], tpr[6], linestyle='--',color='violet', label='Class 6 vs Rest')
plt.plot(fpr[7], tpr[7], linestyle='--',color='pink', label='Class 7 vs Rest')
plt.title('Multiclass ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.savefig('Multiclass ROC',dpi=300);
```
